```python
import numpy as np
import pandas as pd
```

```python
df.drop(columns=['Last Updated', 'Current Ver', 'Android Ver'], inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['Rating'].unique()
```

```python
df['Rating']=df['Rating'].fillna(df['Rating'].mean())
```

```python
df.isnull().sum()
```

```python
df['Type']=df['Type'].fillna(df['Type'].mode()[0])
df['Content Rating']=df['Content Rating'].fillna(df['Content Rating'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Installs'].unique()
```

```python
df['Installs']=df['Installs'].str.replace('[+,]','',regex=True)
df['Installs']=df['Installs'].str.replace('Free','0',regex=False)
df['Installs'].unique()
```

```python
df['Size'].unique()
```

```python
df['Size']=df['Size'].replace('Varies with device',np.nan)
def converter(Size):
  if isinstance(Size,str):
   Size=Size.strip()
   if Size.endswith('M'):
      return float(Size.replace('M',''))
   if Size.endswith('k'):
      return float(Size.replace('k',''))/1024
  return np.nan
df['Size']=df['Size'].apply(converter)
```

```python
df['Size'].unique()
```

```python
df.head()
```

```python
df.isnull().sum()/len(df)*100
```

```python
df['Size']=df['Size'].fillna(df['Size'].median())
```

```python
df.isnull().sum()
```

```python
df['Price'].unique()
```

```python
df['Price']=df['Price'].str.replace('[$]','',regex=True)
df['Price']=df['Price'].str.replace('Everyone','0')
df['Price'].unique()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Price']=df['Price'].astype('float64')
df['Installs']=df['Installs'].astype('Int64')
```

```python
non_numeric_reviews = df[~df['Reviews'].astype(str).str.isnumeric()]
non_numeric_reviews
```

```python
def convert_reviews(x):
    x = str(x)
    if 'M' in x:
        return float(x.replace('M', '')) * 1_000_000
    elif 'K' in x:
        return float(x.replace('K', '')) * 1_000
    elif x.isnumeric():
        return float(x)
    else:
        return np.nan

df['Reviews'] = df['Reviews'].apply(convert_reviews)
```
